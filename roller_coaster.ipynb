# Quais são os locais com as montanhas-russas mais rápidas (mínimo de 10)?

```python
df['Location'].value_counts()
```

```python
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean') ['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Média de velocidade da Montanha-Russa por local')
ax.set_xlabel('Média de velocidade da Montanha-Russa')
ax.set_ylabel('Localização')
plt.show()
```
